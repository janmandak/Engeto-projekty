```
texts = ['''
Situated about 10 miles west of Kemmerer, 
Fossil Butte is a ruggedly impressive 
topographic feature that rises sharply 
some 1000 feet above Twin Creek Valley 
to an elevation of more than 7500 feet 
above sea level. The butte is located just 
north of US 30N and the Union Pacific Railroad, 
which traverse the valley. ''',

'''At the base of Fossil Butte are the bright 
red, purple, yellow and gray beds of the Wasatch 
Formation. Eroded portions of these horizontal 
beds slope gradually upward from the valley floor 
and steepen abruptly. Overlying them and extending 
to the top of the butte are the much steeper 
buff-to-white beds of the Green River Formation, 
which are about 300 feet thick.''',

'''The monument contains 8198 acres and protects 
a portion of the largest deposit of freshwater fish 
fossils in the world. The richest fossil fish deposits 
are found in multiple limestone layers, which lie some 
100 feet below the top of the butte. The fossils 
represent several varieties of perch, as well as 
other freshwater genera and herring similar to those 
in modern oceans. Other fish such as paddlefish, 
garpike and stingray are also present.'''
]

print('-' * 40)
print('Welcome to the app. Please login:')
print('-' * 40)
USERNAME = str(input('USERNAME:'))
PASSWORD = str(input('PASSWORD:'))

credentials = {'bob':'123', 'ann': 'pass123', 'mike': 'password123', 'liz': 'pass123'}
# check of corectness of credentials
while True:
  if USERNAME in credentials and credentials[USERNAME] == PASSWORD:
    print('-' * 40)
    print('Login succesfull.')
    print('-' * 40)
  elif USERNAME not in credentials:
    print('-' * 40)
    print('Sorry, system does not know this user.')
    print('-' * 40)
    break
  else:
    print('-' * 40)
    print('Password is not correct.')
    print('-' * 40)
    break

  print('We have 3 texts to be analyzed.')
  selected_text = int(input('Enter a number btw. 1 and 3 to select: ')) - 1
  print('-' * 40)
  print('You have selected text number', str(selected_text + 1) + '.')
  print('-' * 40)
  print(texts[selected_text])
  print('-' * 40)
  
  # line breaks are causing problem, let's delete them as well as commas and dots
  texts[selected_text] = texts[selected_text].replace('\n', '').replace('.', '').replace(',', '')
  
  # split string by space into list of words
  words = texts[selected_text].split(' ')
  print('Number of words in the selected text: ', len(words))

  # also remove any possible empty list elements
  words.remove('')
  
  counter_capital = 0
  counter_upper = 0
  counter_lower = 0
  counter_digits = 0
  numbers = []
  
  for word in words:
    if word[0].isupper():
      counter_capital += 1
    elif word.isupper():
      counter_upper += 1
    elif word.islower():
      counter_lower += 1
    elif word.isdigit():
      counter_digits += 1
      numbers.append(int(word))
  
  print('Number of titlecase words: ', counter_capital)
  print('Number of uppercase words: ', counter_upper)
  print('Number of lowercase words: ', counter_lower)
  print('Number of numeric strings: ', counter_digits)

  print('-' * 40)

  # frequencies of word lenghts
  lengths = [len(word) for word in words]

  frequencies = {} 
  for item in lengths:
    if (item in frequencies):
      frequencies[item] += 1
    else:
      frequencies[item] = 1
   
  sorted_frequencies = dict(sorted(frequencies.items()))
  
  for freq in sorted_frequencies:
    print(freq, sorted_frequencies[freq]*'*', sorted_frequencies[freq])

  print('-' * 40)

  print('If we summed all the numbers in this text we would get:',  sum(numbers))
  
  break
```
